### Asynchronous displays
The display methods from the displays.py file are called and laucnhed asynchronously from a queue of objects and static parameters.

```python
import asyncio
import threading
from queue import Queue, Empty
import matplotlib.pyplot as plt 
import numpy as np
import os

class Display(threading.Thread):
    def __init__(self, input_queue, function, **static_params):
        super(Display, self).__init__()
        self._input_queue = input_queue 
        self._func = function
        self._statics = static_params
        self.start()
        
    def call_display(self, **params):
        asyncio.run(self._func(**params, **self._statics))
        
    def run(self):
        while True:
            try: 
                value = self._input_queue.get(timeout = 10)
            except Empty:
                break
            if value == "DONE":
                break
            
            params = value
            self.call_display(**params)
            
# parameters
matrix = np.random.random((100,100))
display_queue = Queue()
nb_samples = 4

# test method
async def display_matrix(matrix, matrix_name, cmap = "bwr", outpath = ""):
    print("Calling display")
    plt.figure()
    plt.imshow(matrix, cmap = cmap)
    plt.title(matrix_name)
    plt.savefig(f"{outpath}/{matrix_name}.png")
    
# static parameters
displayer = Display(
    input_queue = display_queue,
    function = display_matrix,
    cmap = "afmhot_r",
    outpath = "displays"
)

if not os.path.exists("displays"):
    os.mkdir("displays")

for i in range(nb_samples):
    display_queue.put({
        "matrix": matrix, 
        "matrix_name": f"test_{ i + 1}"
    })
display_queue.put("DONE")
```

```python
import asyncio
import threading
from queue import Queue, Empty
import matplotlib.pyplot as plt 
import numpy as np
import os
import math

class DisplayBatch(threading.Thread):
    """Same display calls but will batch the inputs with batch size."""
    def __init__(self, input_queue, function, batch_size, **static_params):
        super(DisplayBatch, self).__init__()
        self._input_queue = input_queue 
        self._func = function
        self._batch_size = batch_size
        self._current_batch = []
        self._statics = static_params
        self._nb_batch = 1
        self.start()
        
    def add_to_batch(self, **params):
        """Adds new object params to the current batch"""
        self._current_batch.append(params)
        
    def empty_batch(self):
        self._current_batch = []
        
    def get_current_batch_size(self):
        return len(self._current_batch)
    
    def is_empty_batch(self):
        return self.get_current_batch_size() == 0
        
    def call_batch_display(self):
        asyncio.run(self._func(self._current_batch, title = f"batch#{self._nb_batch}", **self._statics))
        
    def run(self):
        while True:
            try: 
                value = self._input_queue.get(timeout = 10)
            except Empty:
                break
            if value == "DONE":
                break
            
            self.add_to_batch(**value)
            if self.get_current_batch_size() == self._batch_size:
                self.call_batch_display()
                self.empty_batch()
                self._nb_batch += 1
        if not self.is_empty_batch():
            self.call_batch_display()
            
# parameters
matrix = np.random.random((100,100))
display_queue = Queue()
nb_samples = 15
batch_size = 8

# test method
async def display_matrices(matrices, title = "", cmap = "bwr", outpath = ""):
    print("Calling batch display")
    plt.figure()    
    nb_matrices = len(matrices)
    
    cols = math.ceil(math.sqrt(nb_matrices))
    rows = math.ceil(nb_matrices / cols)

    fig, axes = plt.subplots(rows, cols, figsize=(4*cols, 3*rows))
    axes = np.array(axes).reshape(-1)
    for i in range(nb_matrices):
        axes[i].imshow(matrices[i]["matrix"], cmap = cmap)
        axes[i].set_title(f'{matrices[i]["matrix_name"]}')

    # Supprimer les subplots vides (si nb_matrices < rows * cols)
    for j in range(nb_matrices, rows * cols):
        fig.delaxes(axes[j])

    fig.tight_layout()
    plt.savefig(f"{outpath}/{title}.png")
    
# static parameters
displayer = DisplayBatch(
    input_queue = display_queue,
    function = display_matrices,
    batch_size = batch_size,
    cmap = "afmhot_r",
    outpath = "displays"
)

if not os.path.exists("displays"):
    os.mkdir("displays")

for i in range(nb_samples):
    display_queue.put({
        "matrix": matrix, 
        "matrix_name": f"test_{ i + 1}"
    })
display_queue.put("DONE")
```
